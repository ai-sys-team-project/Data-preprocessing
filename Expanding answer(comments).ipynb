```python
def str_to_list(row):
    try:
        return ast.literal_eval(row)
    except (ValueError, SyntaxError) as e:
        print(f"Error parsing row: {row} with error: {e}")
        return []

# 적용
df['comments'] = df['comments'].apply(str_to_list)

# explode 이후 과정
df_exploded = df.explode('comments')

df_exploded['comment_text'] = df_exploded['comments'].apply(lambda x: x['comment'] if isinstance(x, dict) else '')
df_exploded['comment_likes'] = df_exploded['comments'].apply(lambda x: x['likes'] if isinstance(x, dict) else 0)

df_final = df_exploded.drop('comments', axis=1)
df_final.reset_index(inplace=True)
df_final.drop(['index'], axis=1, inplace=True)
```
